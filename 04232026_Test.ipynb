```python
!pip install "numpy<2"

import pandas as pd
import matplotlib.pyplot as plt

# Create simple sample metadata
simple_data = """sample,group,value
A1,Control,23
A2,Control,25
A3,Control,22
B1,Treatment,31
B2,Treatment,29
B3,Treatment,33
C1,Control,24
C2,Treatment,30"""

# Save to CSV file
with open('simple_metadata.csv', 'w') as f:
    f.write(simple_data)

print("Simple CSV file created!")

# Load the CSV
df = pd.read_csv('simple_metadata.csv')

print("Data:")
print(df)
print(f"\nShape: {df.shape}")

# Basic summary stats
print("Summary Statistics:")
print(df['value'].describe())

print("\nGroup counts:")
print(df['group'].value_counts())

print("\nMean by group:")
print(df.groupby('group')['value'].mean())

# Create bar chart
group_means = df.groupby('group')['value'].mean()

plt.figure(figsize=(8, 6))
plt.bar(group_means.index, group_means.values, color=['lightblue', 'lightcoral'])
plt.title('Average Value by Group')
plt.ylabel('Average Value')
plt.xlabel('Group')

# Add value labels on bars
for i, v in enumerate(group_means.values):
    plt.text(i, v + 0.5, f'{v:.1f}', ha='center')

plt.show()
```
